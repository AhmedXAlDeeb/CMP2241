# STACKS STACKS
- is an ordered collection of items where the addition of new items and the removal of existing items always takes place at the same end.
- Items that are closer to the base represent those that have been in the stack the longest.
- Most recently added item is the one that is in position to be removed first.
- Items are ordered based on length of time in the collection.
- Used to reverse the order of items. The order of insertion is the reverse of the order of removal.
- Example: back button in browser actually it is the URLs that are going on the stack).



## Stack implementation As ADT

```python
class Stack:
    def __init__(self):
        self.items = []
    def isEmpty(self):
        return self.items == []
    def push(self, item):
        self.items.append(item) #Add an item to the end of the list. Equivalent to a[len(a):] = [x].
    def pop(self, item):
        return self.items.pop()
    def size():
        return len(self.items)
    def peek(self):
        return self.items[len(self.items)-1]
```

## 1st Problem: Simplfy the Path:
https://leetcode.com/problems/simplify-path/

The input is a string of path. 
first question to ask why would we use a stack?
the url or path is a direct example of stack and Last in First out Principle.

we have a specific string has a sequence of coded operations.
the string is divided into parts by the \
each part is an element or an operation or none
if the part is element then process an opperation of add directlly
if it's an operation process the operation on the last element
the operations can be coded as +1 , +0 , -1
if it's none pass this part

```python
def simplfyPath(path):
    simplePath=[]
    for directory in path.split("/"):
        if directory == "" or directory == ".":
            pass
        elif directory == "..":
            if simplePath: simplePath.pop() 
            else: pass
        else: simplePath.append(directory)
        
    return "/"+ "/".join(simplePath)
```

```python
def simplifyPath(self, path):
    """
    :type path: str
    :rtype: str
    """
    simplePath = []
    directory = ''
    for char in path:
        if char == '/':
            if directory == "" or directory == ".":
                pass
            elif directory == "..":
                if simplePath: simplePath.pop()
            else: simplePath.append(directory)
            directory = ''
        else:
            directory += char
    # handle the last directory if it exists
    if directory:
        if directory == "..":
            if simplePath: simplePath.pop()
        elif directory != ".":
            simplePath.append(directory)
    return "/" + "/".join(simplePath)
```
