## How Linked lists and Trees are related?
 Trees may be stored using a similar kind of structure. If a node in a tree has two children, then that node would have two links to its children as opposed to a linked list which has one  link to the next node in the sequence.

```python
# The binary Search Tree Class
class BST:
    class __Node:
        def __init__(self,val,left=None,right=None):
            self.val= val
            self.left= left
            self.right=right
        def getVal(self):
            return self.val
        def setVal(self,newVal):
            self.val =newVal
        def getLeft(self):
            return self.left
        def setLeft(self, newLeft):
            self.left = newLeft
        def getRight(self):
            return self.right
        def setLeft(self, newRight):
            self.right = newRight

        
```
