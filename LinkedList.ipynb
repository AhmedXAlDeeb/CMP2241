# Linked List Summary and Problems
- There are two simple types of linked structures they are single linked structure and double linked structures.


One direction movement:
-   Singly Linked Structures Uses only one external<B> head </B>link that access the first element >> <i>"it's easily to access the sucessor and hardly the predecessor"</i> 

Double Direction Movement:
-   Double linked structures has an extra link <B> TAil </B> so it's easy to move forward to the sucessor or Backward to the predecessor >> noncontigouse memory
-   the first element has no tail while the last element has no head. >> <B> Empty link</B> >> / 

Memory Allocation:
-   Linked structures arenot accessed by the index although they are accessed by startting at one of the ends then moving towards the elememnt
    -   Insertion and removal done without shifting and During them the structure can be resized easily
- Elements in linked sturctures represnted in a form of node have the value and a head or tail or both.



## implementation of a doubly linked structure:


```python
class Node(): #single direcction node
    def __init__(self, data, next = None):
        """Instantiates a Node with default next of None"""
        self.data = data
        self.next = next
class TwoWayNode(Node): #double direction node
    def __init__(self, data, previous= None , next = None): 
        #Having one extra dummy node at the beginning of the sequence eliminates a lot of special cases
        Node.__init__(self, data, next)
        self.previous = previous

    def getData(self):
        return self.data
    
    def getNext(self):
        return self.next
    
    def setData(self, data):
        self.data = data
        
    def setNext(self, next):
        self.next = next

```
